```python
def d_profil(profil):
    """ Renvoie le nombre de DAG de profil 'profil' """
    produit = (2^1 - 1)^profil[0]
    for k in range(1, len(profil)):
        produit = produit*(2^profil[k-1]-1)^profil[k]
    return produit
```

```python
def dd_profil(profil):
    """ Renvoie le nombre de DAG de profil 'profil' """
    produit = (2^1)^profil[0]
    for k in range(1, len(profil)):
        produit = produit*(2^profil[k-1])^profil[k]
    return produit
```

```python
def repartition_d_profil(n):
    rep = {}
    for e in Compositions(n):
        profil = list(e)
        rep[tuple(profil)] = d_profil(profil)
    return rep


def repartition_dd_profil(n):
    rep = {}
    for e in Compositions(n):
        profil = list(e)
        rep[tuple(profil)] = dd_profil(profil)
    return rep
```

## Programmation dynamique pour calcul des $dd_n$ en fonction de la dernire couche

```python
def nombre_dd_dernier_layer(n,nd):
    """ Renvoie le nombre de DAG dd  n noeuds (racine exclue) avec nd noeuds dans la dernire couche"""
    if n == 0 :
        return 1 if nd==0 else 0
    if nd > n:
        return 0
    if nd == n:
        return 2^n
    
    rep = 0
    for l in range(1, n-nd +1):
        rep += 2^(l*nd)*nombre_dd_dernier_layer(n-nd,l)
        
    return rep


def table_dd_n_nd(N):
    """ Renvoie une table de taille N+1 x N+1 avec en coordonne (n,nd) le nombre de DAG  n noeuds (racine exclue) avec nd noeuds dans la dernire couche"""
    table = [[0 for _ in range(N+1)] for _ in range(N+1)]
    for n in range(N+1):
        table[n][n] = 2^n
    for n in range(2,N+1):
        for nd in range(1,n):
            somme = 0
            for l in range(1,n-nd+1):
                somme += table[n-nd][l]*2^(l*nd)
            table[n][nd] = somme
    return table
            
    
def nombre_dd_dernier_layer_verif(n,nd):
    rep = 0
    dico = repartition_dd_profil(n)
    for k in dico:
        if k[-1] == nd:
            rep += dico[k]
    
    return rep
```

## Programmation dynamique pour calcul des $d_n$ en fonction de la dernire couche

```python
def table_d_n_nd(N):
    """ Renvoie une table de taille N+1 x N+1 avec en coordonne (n,nd) le nombre de DAG  n noeuds (racine exclue) avec nd noeuds dans la dernire couche"""
    table = [[0 for _ in range(N+1)] for _ in range(N+1)]
    for n in range(N+1):
        table[n][n] = 1
    for n in range(2,N+1):
        for nd in range(1,n):
            somme = 0
            for l in range(1,n-nd+1):
                somme += table[n-nd][l]*(2^l - 1)^nd
            table[n][nd] = somme
    return table
```
