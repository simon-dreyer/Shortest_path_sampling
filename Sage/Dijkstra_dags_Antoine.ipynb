```python
var('x, y')
N = 70
```

```python
F0 = taylor(taylor(x*y / (1-x*y),x , 0, N), y, 0, N)
F0
```

```python
F = F0
for l in range(N):
    F = F0 + sum([taylor(taylor(expand((x*y)**k * F.subs(y = 2**k-1)),x , 0, N), y, 0, N) for k in range(1,N)])
```

```python
f = F.subs(y=1)
f
```

```python
G0 = taylor(taylor(x*y / (1-x*y),x, 0, N), y, 0, N)
#print(G)
G = G0
for l in range(N):
    G = G0  + sum([taylor(taylor(expand((x*y)**k * G.subs(y = 2**k)),x , 0, N), y, 0, N) for k in range(1,N)])
   # print(G.subs(y=1))
```

```python
G
```

```python
g = G.subs(y=1)
g
```

```python
point([(i, g.coefficient(x, i)/f.coefficient(x, i)) for i in range(1,N)])
```

```python
[f.coefficient(x, k) for k in range(10)]
```

```python
[g.coefficient(x, k) for k in range(10)]
```

```python
var('x,y,z')
N = 30

H0 = taylor(taylor(x*y / (1-x*y),x, 0, N//2), y, 0, N)

H = H0
for l in range(N):
    H = expand( taylor(taylor(taylor(expand(z*x*(H.subs(y = 2*y) - H) + H.subs(y=1).subs(z=y)),x , 0, N//2), y, 0, N), z, 0, N))
   
```

```python
h = H.subs(y=1).subs(z=1)
```

```python
[f.coefficient(x, k) for k in range(16)] == [h.coefficient(x, k) for k in range(16)] 
```
